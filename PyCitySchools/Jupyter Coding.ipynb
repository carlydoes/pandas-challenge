```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Total number of unique schools
total_school = school_data["school_name"].count()
total_school
```

```python
#Total students
total_students = student_data["student_name"].count()
total_students
```

```python
#Total budget 
total_budget = school_data["budget"].sum()
total_budget

#Average math score
avg_score = student_data["math_score"].mean()
avg_score

#Average reading score
avg_reading = student_data["reading_score"].mean()
avg_reading

# % passing math
passing_math = (school_data_complete["math_score"]>70).sum()/school_data_complete["math_score"].count() *100
passing_math

# % passing reading
passing_reading = (school_data_complete["reading_score"]>70).sum()/school_data_complete["reading_score"].count() *100
passing_reading

# % overall passing
overall_passing = ((avg_score + avg_reading) /2)
overall_passing
```

```python
#District Summary
Summary = pd.DataFrame ({"total_school":[total_school], "total_students":[total_students],"total_budget":[total_budget],"avg_score":[avg_score],"avg_reading":[avg_reading],"passing_math":[passing_math],"passing_reading":[passing_reading], "overall passing":[overall_passing]})
Summary
```

```python
#total students
student_total = school_data_complete["school_name"].value_counts()
student_total.head()
```

```python
#average math score per school
avg_math_per = school_data_complete.groupby(["school_name"])["math_score"].mean()
avg_math_per.head()
```

```python
# % passing math (the percentage of students who passed math)
passing_math_per = school_data_complete.loc[school_data_complete["math_score"]>=70]
passing_math_per

group = passing_math_per["school_name"].value_counts()
group.head()

percent_math = group/student_total * 100
percent_math
```

```python
# % passing reading (the percentage of students who passed reading)
passing_reading_per = school_data_complete.loc[school_data_complete["reading_score"]>=70]
passing_reading_per

group_reading = passing_reading_per["school_name"].value_counts()
group_reading.head()

percent_reading = group_reading/student_total * 100
percent_reading
```

```python
# % overall passing (the percentage of students who passed math AND reading)
overall = percent_math + percent_reading / total_students
overall
```

```python
# School Summary 
school_summary_df = pd.DataFrame({"School Type":school_types, "Total Students":total_students, "Total School Budget":school_budget
                                ,"Per Student Budget":per_budget ,"Average Math Score" :avg_math_per,
                             "Average Reading Score":avg_reading_per,
                             "% Passing Math":percent_math, "% Passing Reading":percent_reading,
                            "% Overall Passing Rate":overall})
school_summary_df.head()
```
